### Grouping

**df.groupby(by=grouping_columns)[columns_to_show].function()**

### Summary Tables

Suppose we want to see how the observations in our sample are distributed in the context of two variables - Churn and International plan.\
To do so, we can build a contingency table using the crosstab method:

```python
pd.crosstab(df["Churn"], df["International plan"])
```

```python
pd.crosstab(df["Churn"], df["International plan"], normalize=True)
```

### Pivot table

* values  a list of variables to calculate statistics for,
* index  a list of variables to group data by,
* aggfunc  what statistics we need to calculate for groups, ex. sum, mean, maximum, minimum or something else.

```python
df.pivot_table(columns_to_show,["Churn"], aggfunc="mean")
```

## 2. First attempt at predicting telecom churn

Let's see how Churn rate is related to International plan feature

```python
pd.crosstab(df["International plan"], df["Churn"], margins=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Graphics in retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
```

```python
sns.countplot(data=df, x="International plan", hue= "Churn")
```

Now lets look into the Customer Service calls feature

```python
pd.crosstab(df["Customer service calls"], df["Churn"], margins=True).T
```

```python
sns.countplot(data=df, x="Customer service calls", hue = "Churn")
```

Though is not highly related, if you look after 4 calls, there's an increase in Churn rate.
Lets create a feature -> customer service call > 3

```python
df["Customer service calls > 3"] = df["Customer service calls"].apply(lambda x : x>3)
df["Customer service calls > 3"]
```

```python
pd.crosstab(df["Customer service calls > 3"], df["Churn"], margins=True)
```

```python
sns.countplot(data = df, x = "Customer service calls > 3", hue = "Churn")
```

There is good relation for this feature

Let's create a crosstab which correlated "International plan" and Customer service calls > 3"

```python
pd.crosstab(df["International plan"] & df["Customer service calls > 3"], df["Churn"], margins=True)
```
